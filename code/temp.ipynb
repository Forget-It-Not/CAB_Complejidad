```python
import numpy as np
```

```python
def model(a=0, b=0):
    return a + b
```

```python
a = np.random.uniform(-5,5,size=1000)
b = np.random.uniform(-5,5,size=1000)
y = model(a=a, b=b)

df = pd.DataFrame({'a':a,'b':b,'y':y})
```

```python
fig = px.scatter_3d(df, x='a',y='b',z='y')
fig.show()
```

```python
df_pca = PCA(n_components=2).fit_transform(df)
df_pca
```
