```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import poisson
```

Función que da el promedio de uniones por unidad de tiempo.

El promedio máximo se da con $N(R) = N_N$. Asumiendo que se establece $V = \alpha N_N$, dicha cantidad sería $k N_N / \alpha$. 

Para que el promedio sea "válido" lo único que tiene que cumplir es que de lugar a una cantidad de uniones $\le N(R)/2$ (en el caso máximo, $\le N_N/2$). Si se combinan las constantes $k$ y $\alpha$ como $1/2$, el promedio queda dentro de los límites razonables pero hay una probabilidad sustancial (~50%) de que la cantidad concreta muestreada no lo cumpla. 

$\Rightarrow$ combinamos $k$ y $\alpha$ como $1/4$

```python
def freq_union(nn, V):
    # k * N(R)^2 / V
    return (nn**2) / (4*V)
```

```python
NN=40
nnet = np.arange(1,NN+1)
fu = freq_union(nnet, NN)

nun = np.arange(1,NN/2 +1)
FU,NUn = np.meshgrid(fu,nun)
P = poisson.pmf(NUn,FU)

plt.plot(nnet,fu,'--',color='black')
plt.imshow(P, cmap='Reds')
IC0,IC1 = poisson.interval(0.95, fu)
plt.plot(nnet, IC0, color='black', label='IC 95%')
plt.plot(nnet, IC1, color='black')
plt.colorbar()
plt.ylim(0,20)
plt.legend()
```
